```python
#Q9 Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    plt.hist(z_scores, bins=30, density=True, color='orange', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

    return z_scores

data = np.random.normal(50, 10, 1000)  # Simulated dataset
z_scores = calculate_and_plot_zscores(data)
```

```python
#Q10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt
population_size = 300000
sample_size = 35
num_samples = 1000
population = np.random.exponential(scale=2.0, size=population_size)
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))
plt.hist(sample_means, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Central Limit Theorem Demo\nDistribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#Q14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

def plot_standard_normal_with_z(z_score):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)
    plt.plot(x, y, label='Standard Normal PDF', color='blue')
    plt.fill_between(x, y, where=(x <= z_score), color='lightblue', alpha=0.5)
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
    plt.title('Z-score on Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True)
    plt.legend()
    plt.show()

data_point = 80
population_mean = 75
population_std = 10

z = calculate_z_score(data_point, population_mean, population_std)
print(f"Z-score: {z:.2f}")

# Plot the Z-score on standard normal distribution
plot_standard_normal_with_z(z)
```

```python
#Q Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm
sample = np.array([75, 78, 78, 71, 69, 74, 73, 70, 76, 77])
population_mean = 71
population_std = 5
sample_mean = np.mean(sample)
n = len(sample)

z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z_stat:.2f}")
print(f"P-value = {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
#Q15 Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

# Sample dataset (e.g., heights, scores, etc.)
data = np.array([72, 75, 78, 71, 69, 74, 73, 70, 76, 77])

# Confidence level
confidence = 0.95

# Calculate statistics
n = len(data)
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error of the mean

# Margin of error
margin = stats.t.ppf((1 + confidence) / 2, df=n-1) * std_err

# Confidence interval
ci_lower = mean - margin
ci_upper = mean + margin

print(f"Sample Mean = {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q16 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
np.random.seed(45)
data = np.random.normal(loc=100, scale=25, size=70)
mean = np.mean(data)
sem = stats.sem(data)
confidence = 0.95
df = len(data) - 1
t_critical = stats.t.ppf((1 + confidence) / 2, df)
margin_of_error = t_critical * sem
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error
print(f"Sample Mean = {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', label='Sample Mean')
plt.axvspan(ci_lower, ci_upper, color='yellow', alpha=0.3, label='95% Confidence Interval')
plt.title('Histogram with 95% Confidence Interval for Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q23 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=2000):
    data = np.random.normal(loc=mean, scale=std, size=size)
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)

    print(f"Skewness: {data_skewness:.4f} (≈ 0 for symmetric distribution)")
    print(f"Kurtosis: {data_kurtosis:.4f} (≈ 0 for normal distribution)")

    return data_skewness, data_kurtosis

analyze_normal_distribution_skew_kurtosis()
```
